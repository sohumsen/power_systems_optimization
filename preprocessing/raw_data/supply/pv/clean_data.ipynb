```python
all_dates=[d.strftime('%d/%m/%Y %H:%M:%S') for d in pd.date_range("01/01/2018", periods=365*24, freq="H")]

# 01/01/2018  00:00:00
print(all_dates)
```

```python
df=pd.read_excel("solar.xls",)



def get_power_at_time_for_county(date_time,county):
    print(date_time,county)
    index=df.index
    condition = df[county+" time"]==date_time
    apples_indices = index[condition]
    apples_indices_list = apples_indices.tolist()
    if (len(apples_indices)==0):
        return 0
    for i in range(len(list(df.columns.values))):

        column_name=list(df.columns.values)[i]
        if county in column_name:

                
            if "power" in column_name:


                return df[df.columns[i]][apples_indices_list[0]]


def get_clean_county_names():

    c= [x.replace('power', '').replace('time', '') for x in list(df.columns.values)]
    return c
get_power_at_time_for_county("01/01/2018 11:00","Wiltshire-Lyneham")
```

```python
import csv

county_names=["Wiltshire-Lyneham","Goucestershire-Little Rissington","Cornwall-Camborne","Devon-Exeter" ,"Cornwall-Cardinham Bodmin","Somerset-Liscombe","Somerset-Yeovilton","Dorset-Hurn","Hampshire-Odiham","Surrey-Wisley","Kent-East Malling","Kent-Manston" ,"Greater London-Heathrow","Greater London-Kew Gardens","Greater London-Kenley"]
# county_names=get_clean_county_names()
for i in range(len(all_dates)):

    lst=[all_dates[i]]

    for j in range(len(county_names)):


        b=get_power_at_time_for_county(all_dates[i],county_names[j])
        lst.append(b)

    with open('employee_file.csv', mode='a') as employee_file:
        employee_writer = csv.writer(employee_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        employee_writer.writerow(lst)

```
